```python
url = 'https://kuban.rbc.ru/'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64; rv:107.0) Gecko/20100101 Firefox/107.0',
}
answer = requests.get(url, headers=headers)
```

```python
dom = html.fromstring(answer.text)
link_list = dom.xpath('//span[@class="main__feed__title"]/text()')
```

```python
n = 1
for el in link_list:
    print(n, '. ',el, end='\n')
    n += 1
```
